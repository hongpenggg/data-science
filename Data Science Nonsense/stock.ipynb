```python
# The data

from datetime import datetime

# Read the CSV file
df = pd.read_csv('AAPL5Y.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the number of days since '05/22/2019' for each date
start_date = datetime.strptime('05/22/2019', '%m/%d/%Y')
df['Days'] = (df['Date'] - start_date).dt.days

# Extract the 'Close/Last' prices and remove the '$' sign
df['Close/Last'] = df['Close/Last'].str.replace('$', '').astype(float)

# Now, 'X' is the 'Days' column and 'Y' is the 'Close/Last' column
X = df['Days'].values.reshape(-1, 1)
Y = df['Close/Last'].values.reshape(-1, 1)

print(X)
print(Y)
```

```python
# getting the test data

import pandas as pd
from datetime import datetime

# Read the CSV file
ndf = pd.read_csv('AAPLTest.csv')

# Convert the 'Date' column to datetime format
ndf['Date'] = pd.to_datetime(ndf['Date'])

# Calculate the number of days since '05/22/2019' for each date
start_date = datetime.strptime('05/22/2019', '%m/%d/%Y')
ndf['Days'] = (ndf['Date'] - start_date).dt.days

# Now, 'X' is the 'Days' column
x_test = ndf['Days'].tolist()

x_test = np.array(x_test).reshape(-1, 1)

print(x_test)
```

```python
# Transform X to polynomial features
poly = PolynomialFeatures(degree=94)
X_poly = poly.fit_transform(X)

# Fit the model to the data
reg = LinearRegression()
reg.fit(X_poly, Y)

# Transform x_test to polynomial features
x_test_poly = poly.transform(x_test)

# Predict the value for test data
prediction = reg.predict(x_test_poly)

print("The prediction for the test data is:", prediction+20)
```
