```python
# Normalize the input features

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(df['Days'].values.reshape(-1, 1))
Y = df['Close/Last'].values.reshape(-1, 1)

print(X)
print()
print(Y)
```

```python
# importing tensorflow dependencies

%pip install tensorflow

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense

from sklearn.metrics import accuracy_score
```

```python
'''
# build and compile tensorflow model

model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=1))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
'''

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# build and compile tensorflow model
model = Sequential()
model.add(Dense(units=256, activation='relu', input_dim=1))
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=256, activation='relu'))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=1))  # Remove 'sigmoid' activation for regression
```

```python
# model.compile(loss='mean_squared_error', optimizer='sgd')


# compile model

optimizer = Adam(learning_rate=0.0005)
model.compile(loss='mean_squared_error', optimizer=optimizer)
```

```python
# fit model

# model.fit(X, Y, epochs=500, batch_size=40)


# fit model
early_stop = EarlyStopping(monitor='val_loss', patience=20)
# model.fit(X, Y, epochs=1000, batch_size=8, validation_split=0.2, callbacks=[early_stop])

model.fit(X, Y, epochs=8000, batch_size=4, validation_split=0.2)
```

```python
# getting the test data
'''
import pandas as pd
from datetime import datetime

# Read the CSV file
ndf = pd.read_csv('AAPLTest.csv')

# Convert the 'Date' column to datetime format
ndf['Date'] = pd.to_datetime(ndf['Date'])

# Calculate the number of days since '05/22/2019' for each date
start_date = datetime.strptime('05/22/2019', '%m/%d/%Y')
ndf['Days'] = (ndf['Date'] - start_date).dt.days

# Now, 'X' is the 'Days' column
x_test = ndf['Days'].tolist()

x_test = np.array(x_test).reshape(-1, 1)

print(x_test)
'''

# getting the test data
ndf = pd.read_csv('AAPLTest.csv')
ndf['Date'] = pd.to_datetime(ndf['Date'])
ndf['Days'] = (ndf['Date'] - start_date).dt.days
x_test = scaler.transform(ndf['Days'].values.reshape(-1, 1))
y_test = ndf['Close/Last'].str.replace('$', '').astype(float).values.reshape(-1, 1)

print(x_test)
print(y_test)
```

```python
# evaluate model

y_hat = model.predict(x_test)
print(y_hat)
```

```python
# Calculate the evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score

y_true = y_test
y_pred = y_hat

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"Root Mean Squared Error: {rmse}")

r2 = r2_score(y_true, y_pred)
print(f"R-squared: {r2}")
```
